```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

np.random.seed(42)
tf.random.set_seed(42)
```

```python
from google.colab import files

uploaded = files.upload()
fname = list(uploaded.keys())[0]

if fname.endswith(".csv"):
    df = pd.read_csv(fname)
else:
    df = pd.read_excel(fname)

df.head()
```

```python
# 1) Check symbol column
if "symbol" not in df.columns:
    raise ValueError("Expected a 'symbol' column in the dataset.")

# 2) Symbol → Sector mapping (extend if needed)
sector_map = {
    # Information Technology
    "AAPL": "Information Technology",
    "MSFT": "Information Technology",
    "NVDA": "Information Technology",
    "AVGO": "Information Technology",
    "ADBE": "Information Technology",

    # Communication Services
    "GOOGL": "Communication Services",
    "GOOG":  "Communication Services",
    "META":  "Communication Services",
    "NFLX":  "Communication Services",

    # Consumer Discretionary
    "AMZN": "Consumer Discretionary",
    "TSLA": "Consumer Discretionary",
    "HD":   "Consumer Discretionary",
    "MCD":  "Consumer Discretionary",

    # Financials
    "JPM":  "Financials",
    "BAC":  "Financials",
    "BRK.B":"Financials",
    "BRK.A":"Financials",
    "V":    "Financials",
    "MA":   "Financials",
    "WFC":  "Financials",

    # Health Care
    "UNH":  "Health Care",
    "LLY":  "Health Care",
    "JNJ":  "Health Care",
    "ABBV": "Health Care",
    "PFE":  "Health Care",

    # Consumer Staples
    "PG":   "Consumer Staples",
    "KO":   "Consumer Staples",
    "PEP":  "Consumer Staples",
    "COST": "Consumer Staples",
    "WMT":  "Consumer Staples",

    # Energy
    "XOM": "Energy",
    "CVX": "Energy",
    "COP": "Energy",

    # Industrials (examples)
    "BA":  "Industrials",
    "CAT": "Industrials",
    "GE":  "Industrials",
}

# 3) Create Sector column
df["Sector"] = df["symbol"].map(sector_map).fillna("Other")

print(df[["symbol", "Sector"]].drop_duplicates().head(20))
print("Unique sectors:", df["Sector"].unique())
```

```python
TARGET_COL = "freeCashFlow(in billions)"
TIME_COL   = "quarter"   # change if your time column is named differently
SECTOR_COL = "Sector"

feature_cols = [
    "netCashProvidedByOperatingActivities(in billions)",  # micro 1
    "netCashUsedForInvestingActivites(in billions)",      # micro 2
    "cashAndCashEquivalents(in billions)",                # micro 3
    "revenue(in billions)",                               # micro 4
    "operatingIncome(in billions)",                       # micro 5
    "totalLiabilities(in billions)",                      # micro 6
    "totalAssets(in billions)",                           # micro 7
    "GDP_Growth",                                         # macro 1
    "Interest_Rate",                                      # macro 2
    "SP500_Growth",                                       # macro 3
]

cols_needed = [TARGET_COL, SECTOR_COL] + feature_cols
df_q = df.dropna(subset=cols_needed).copy()

print("After cleaning:", df_q.shape)
print("Sectors in cleaned data:", df_q[SECTOR_COL].unique())
```

```python
def build_lstm_model(n_features):
    model = Sequential()
    model.add(LSTM(32, activation='tanh', return_sequences=False, input_shape=(1, n_features)))
    model.add(Dense(16, activation='tanh'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model
```

```python
results_lstm = []
PLOT_PER_SECTOR = True

sectors = df_q[SECTOR_COL].dropna().unique()
num_features = len(feature_cols)

print("Sectors detected for LSTM:", sectors)

for sec in sectors:
    print("\n============================")
    print(f"Processing Sector (LSTM): {sec}")
    print("============================")

    sub = df_q[df_q[SECTOR_COL] == sec].copy()
    n_rows = len(sub)
    print(f"Rows for {sec}: {n_rows}")

    if n_rows < 12:
        print(f"⚠ Too few rows for LSTM in {sec}. Skipping.")
        continue

    # Sort by time if present
    if TIME_COL in sub.columns:
        sub = sub.sort_values(TIME_COL)

    # Build X & y
    X = sub[feature_cols].values
    y = sub[TARGET_COL].values

    # Scale features + target
    X_scaler = StandardScaler()
    y_scaler = StandardScaler()

    X_scaled = X_scaler.fit_transform(X)
    y_scaled = y_scaler.fit_transform(y.reshape(-1,1)).ravel()

    # LSTM requires 3D input: (samples, timesteps=1, features)
    X_lstm = X_scaled.reshape((X_scaled.shape[0], 1, num_features))

    # 80/20 train-test split (time-based)
    split_idx = max(5, int(0.8 * len(X_lstm)))
    X_train, X_test = X_lstm[:split_idx], X_lstm[split_idx:]
    y_train, y_test = y_scaled[:split_idx], y_scaled[split_idx:]

    if len(X_test) == 0:
        print(f"⚠ No test samples for {sec}. Skipping.")
        continue

    # Build and train LSTM
    print(f"Training LSTM for {sec} ...")
    model = build_lstm_model(num_features)
    model.fit(X_train, y_train, epochs=50, batch_size=8, verbose=0)

    # Predict
    y_pred_scaled = model.predict(X_test).ravel()

    # Inverse scaling
    y_pred_real = y_scaler.inverse_transform(y_pred_scaled.reshape(-1,1)).ravel()
    y_test_real = y_scaler.inverse_transform(y_test.reshape(-1,1)).ravel()

    # Metrics
    mae  = mean_absolute_error(y_test_real, y_pred_real)
    #rmse = mean_squared_error(y_test_real, y_pred_real, squared=False)
    r2   = r2_score(y_test_real, y_pred_real)

    print(f"✔ LSTM Metrics for {sec}: MAE={mae:.3f}, R2={r2:.3f}")
    #RMSE={rmse:.3f},

    results_lstm.append({
        "Sector": sec,
        "Rows": n_rows,
        "MAE": mae,
        #"RMSE": rmse,
        "R2": r2
    })

    # Plot last 20 points
    if PLOT_PER_SECTOR:
        N = min(20, len(y_test_real))
        plt.figure(figsize=(8,4))
        plt.plot(y_test_real[-N:], label="Actual", marker="o")
        plt.plot(y_pred_real[-N:], label="Predicted (LSTM)", marker="x", linestyle="--")
        plt.title(f"LSTM Regression — {sec} (Last {N} Test Points)")
        plt.xlabel("Index in Test Set")
        plt.ylabel("Free Cash Flow (in billions)")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()
```

```python
results_lstm_df = pd.DataFrame(results_lstm).sort_values("Sector")
results_lstm_df
```

```python
results_lstm_df.to_csv("Sectorwise_LSTM_Results.csv", index=False)
```
